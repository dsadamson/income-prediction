```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
```

```python
df = pd.read_csv("income.csv")
```

```python
df.info()
```

```python


# Analyze distribution of numerical variables within income categories using histograms
numerical_vars = ['age', 'hours_per_week']
for var in numerical_vars:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x=var, hue='income', multiple='stack', kde=True)
    plt.title(f"Distribution of {var} by Income")
    plt.xlabel(var)
    plt.ylabel("Count")
    plt.legend(["Greater Than 50K","Less than or equal to 50K"],title="Income", loc='upper right', fontsize = "small")
    plt.savefig(f"{var}.png")
    plt.show()
```
